#### Task 1: Importing libraries

```python
import numpy as np 
import cv2 
import matplotlib.pyplot as plt
```

#### Task 2: Read the Image and make a grid

```python
img=cv2.imread('highway.png')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
print("shape of image",img.shape)
plt.grid()
plt.imshow(img)
```

#### Task3: Build a Bird View for image

```python
pts1 = np.float32([[758,800],[1059,800],[250,1000],[1750,1000]])
pts2 = np.float32([[0,0],[500,0],[0,600],[500,600]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(500,600))
plt.figure(figsize=(16, 12))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

#### Task4 :Read image and Build a Bird View for video

```python
cap =cv2.VideoCapture('road.mp4')
```

```python
while(1):
    ret,frame = cap.read()  # reading each frame : ret is a boolean if reading is done correctly or not , frame
    pts1 = np.float32([[758,800],[1059,800],[250,1000],[1750,1000]])
    pts2 = np.float32([[0,0],[500,0],[0,600],[500,600]])
    M = cv2.getPerspectiveTransform(pts1,pts2)
    dst = cv2.warpPerspective(img,M,(500,600))
    cv2.imshow('road',frame)    
    cv2.imshow('Bird Eye View',dst)    
    if cv2.waitKey(1000) & 0xFF == ord('q'):# 0xFF is a hexadecimal constant which is 11111111 in binary.
        break                             # By using bitwise AND (&) with this constant,                                               
cap.release()                             # it leaves only the last 8 bits of the original
cv2.destroyAllWindows()                   #(in this case, whatever cv2.waitKey() is)                     
```
